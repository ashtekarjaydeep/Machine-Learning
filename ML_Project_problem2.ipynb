# 2.Remove punctuations and digits

```python
# remove punctuation
for i in range(len(df)):
    df[i]['sentence'] = df[i]['sentence'].str.replace('[^\w\s]','')
    df[i].head()
```

```python
# remove digits
for i in range(len(df)):
    df[i]['sentence']=[x for x in df[i]['sentence'] if not x.isdigit()]
    
```

```python
#lower case
for i in range(len(df)):
    df[i]['sentence'] = df[i]['sentence'].apply(lambda x: " ".join(x.lower() for x in x.split()))
    df[i].head()
```

```python
# stop words removal
from nltk.corpus import stopwords
stop = stopwords.words('english')
for i in range(len(df)):
    df[i]['sentence'] = df[i]['sentence'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
    df[i]['sentence'].head()
```

```python
for i in range(len(df)):
    df[i]['totalwords'] = [len(x.split()) for x in df[i]['sentence'].tolist()]
    df[i][['sentence','totalwords']].head()
```

```python
for i in range(len(df)):
    df[i]['char_count'] = df[i]['sentence'].str.len() ## this also includes spaces
    df[i][['sentence','char_count']].head()
```

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
for i in range(len(df)):
    df[i]['sentence'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

    df[i].head()
```

```python
from wordcloud import WordCloud,STOPWORDS
new_df=df_Nixon['sentence']
words = ' '.join(new_df)
cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and word != 'RT'
                            ])
wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(cleaned_word)
plt.figure(1,figsize=(12, 12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
